```python
# make plot of performance w.r.t. ratio of data used
baselines_ls = [baseline_1, baseline_10, baseline_50, baseline_100]
tangos_01_001_ls = [
    tangos_1_01_001,
    tangos_10_01_001,
    tangos_50_01_001,
    tangos_100_01_001,
]
tangos_1_01_ls = [tangos_1_1_01, tangos_10_1_01, tangos_50_1_01, tangos_100_1_01]
tangos_1_001_ls = [tangos_1_1_001, tangos_10_1_001, tangos_50_1_001, tangos_100_1_001]
tangos_10_01_ls = [tangos_1_10_01, tangos_10_10_01, tangos_50_10_01, tangos_100_10_01]
l2_00001_ls = [l2_1_00001, l2_10_00001, l2_50_00001, l2_100_00001]
l2_0001_ls = [l2_1_0001, l2_10_0001, l2_50_0001, l2_100_0001]


data_dict = {"Baseline": baselines_ls, "TANGOS": tangos_1_001_ls, "L2": l2_00001_ls}

col_dict = {"Baseline": "dimgrey", "TANGOS": blue, "L2": green}


fontsize = 20
fs_legend = 17
plt.figure(figsize=(10, 5))
# plt.title('Dionis')
plt.ylabel("\% Test Accuracy", fontsize=fontsize)
plt.xlabel("\% Training Data Used", fontsize=fontsize)
x_vals = [0, 50, 100]
plt.xticks(x_vals, ["10", "50", "100"], fontsize=fontsize)
y_vals = [0.76, 0.8, 0.84, 0.88]
plt.yticks(y_vals, [76, 80, 84, 88], fontsize=fontsize)
plt.ylim(0.76 - 0.01, 0.88 + 0.005)
for method, results in data_dict.items():
    results = [proportion[:6] for proportion in results]  # all should have equal runs
    means = [np.mean(proportion) for proportion in results]
    stds = [np.std(proportion) for proportion in results]
    means = means[1:]
    stds = stds[1:]
    upper = [m + s for m, s in zip(means, stds)]
    lower = [m - s for m, s in zip(means, stds)]
    # x_vals = [10, 50, 100]
    if method != "Baseline":
        plt.plot(
            x_vals,
            means,
            c=col_dict[method],
            label=method,
            linewidth=3.5,
            linestyle="-",
            marker="s",
        )
        plt.fill_between(x_vals, upper, lower, color=col_dict[method], alpha=0.25)
    else:
        plt.plot(
            x_vals,
            means,
            c=col_dict[method],
            label=method,
            linewidth=3.5,
            linestyle="-",
        )
handles, labels = plt.gca().get_legend_handles_labels()
order = [1, 2, 0]
lgnd = plt.legend(
    [handles[idx] for idx in order],
    [labels[idx] for idx in order],
    fontsize=fs_legend,
    frameon=False,
)
for handle in lgnd.legendHandles:
    handle.set_linewidth(5)
plt.savefig("scale.pdf", format="pdf", dpi=1200)
plt.show()
```
